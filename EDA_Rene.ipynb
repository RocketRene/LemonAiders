```python

import pandas as pd
import uuid
import numpy as np
```

```python
monday = pd.read_csv('data/monday.csv',sep=';',parse_dates=['timestamp'])
tuesday = pd.read_csv('data/tuesday.csv',sep=';',parse_dates=['timestamp'])
wednesday = pd.read_csv('data/wednesday.csv',sep=';',parse_dates=['timestamp'])
thursday = pd.read_csv('data/thursday.csv',sep=';',parse_dates=['timestamp'])
friday = pd.read_csv('data/friday.csv',sep=';',parse_dates=['timestamp'])
```

```python
def create_id(chunck):
    chunck['id']=uuid.uuid1()
    return chunck
    
```

```python

for name, group in monday.groupby('customer_no'):
    
    print('\n')
    
    print(name)
    print(group)
    print('\n')
    #print()
    print('\n')
    group['id']=uuid.uuid1()
    print(group['id'])
    
```

```python
dfs =[monday,tuesday,wednesday,thursday,friday]
```

```python
out = []
for frame in dfs:
    
    frame= frame.groupby('customer_no').apply(create_id)
    print(frame.shape)
    out.append(frame)
```

```python
df= pd.DataFrame()
```

```python
df = pd.concat(out)
```
